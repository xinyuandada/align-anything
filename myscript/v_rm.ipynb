```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re

#  JSON 
with open('/data/zhangyanhong-2401220256/align-anything/outputs/qwen_2_5_rm_save_score/5_20_one-epoch/tmp/process_0.json', 'r') as f:
    data = json.load(f)

#  h_r 
h_r_list =  [item["h_r"] for item in data]  # 
l_r_list =  [item["l_r"] for item in data]   # 
margin_r_list =  [item["h_r"]-item["l_r"] for item in data]   # 

print(margin_r_list)


# 
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")  # 35

#  + KDE
sns.histplot(h_r_list, kde=True, color='#2196F3', 
             edgecolor='white', alpha=0.5, line_kws={'lw': 2},label='dpo model')

#  + KDE
sns.histplot(l_r_list,  kde=True, color='red', 
             edgecolor='white', alpha=0.5, line_kws={'lw': 2},label='origin model')

#  + KDE
#sns.histplot(margin_r_list,  kde=True, color='yellow', edgecolor='white', alpha=0.5, line_kws={'lw': 2})
plt.legend(loc='upper right', fontsize=10, frameon=True, shadow=True)
# 
plt.title("Distribution of score Values", fontsize=14, pad=15)
plt.xlabel("score", fontsize=12)
plt.ylabel("Density/Frequency", fontsize=12)
#plt.xticks(np.arange(-3, 0, 0.5))  # 15
#plt.xlim(-2.5, -0.5)  # 

# 
plt.show()

    
```

```python
import numpy as np
import matplotlib.pyplot as plt

origin_score = []
origin_length = []
dpo_score = []
dpo_length = []
pattern = re.compile(r'assistant\.\nuser\n(.*)', re.DOTALL) 

for item in data:
    dpo_score.append(item["h_r"])
    origin_score.append(item["l_r"])
    dpo_length.append(len(pattern.search(item["h_c"]).group(1)))     
    origin_length.append(len(pattern.search(item["l_c"]).group(1)))
    
        
# 12
coefficients = np.polyfit(dpo_length, dpo_score, deg=2)  # deg=1 deg=2 
trend_line = np.polyval(coefficients, dpo_length)

# 
plt.figure(figsize=(10, 6))
plt.scatter(dpo_length, dpo_score, color='#2196F3', alpha=0.6, s=50, label='Data Points')
plt.plot(dpo_length, trend_line, color='#ee2500', alpha=0.6, lw=2, label='Trend Line')

# 
plt.title("DPO Model", fontsize=14, pad=15)
plt.xlabel("response length", fontsize=12)
plt.ylabel("score", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()


```
